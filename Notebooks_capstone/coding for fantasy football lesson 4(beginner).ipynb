```python
import pandas as pd
```

```python
csv_path = '2019.csv'
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
df = df.loc[df['Pos'] == 'RB', ['Player', 'Tgt', 'RushingAtt', 'FantasyPoints']]
df['Usage'] = df['Tgt'] + df['RushingAtt']
```

ok there is alot going on so here goes a breakdown of this code
_________________________________________________________________

.loc lets us select portions of our dataframe based off of the row then column in our dataframe.

                .loc[(df['Pos'] == 'RB')
                
this first portion is our row selector, what we are telling python is that for each for in our df only choose the ones that the 'Pos' is equal to 'RB'. This is called Boolean indexing




the second part of our .iloc is the column selector. in this part you are telling pandas these are the columns that i want in my dataframe. 

               , ['Player', 'Tgt', 'RushingAtt', 'FantasyPoints']]
  
 This way we only keep these columns and exclude everything else from our data frame.

df['Usage'] = df['Tgt'] + df['RushingAtt']

ok so this part of the code is pretty intuitive to me at this point basically it is saying the new column in df is called ['Usage'] and usage is equal to df['Tgt'] + df['RushingAtt']
