```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
csv_path = '2019.csv'
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
df = df.loc[df['Pos'] == 'RB', ['Player', 'Tgt', 'RushingAtt', 'FantasyPoints']]
df['Usage'] = df['Tgt'] + df['RushingAtt']
```

```python
df['UsageRank'] = df['Usage'].rank(ascending = False)
df['FantasyPointsRank'] = df['FantasyPoints'].rank(ascending = False)
df.sort_values(by = 'UsageRank').head(15)
```

```python
def covariance(x, y):
    n = len(x)
    return sum((x - np.mean(x)) * (y - np.mean(y))) * 1/(n-1)
```

```python
x = df['Usage'].values
y = df['FantasyPoints'].values
```

```python
def corr(x, y):
    return covariance(x, y)/(np.std(x) * np.std(y))
corr(x, y)
```

Machine learning algorithms are classified in to two categories- those that "train" or learn with data that is labeled and those that dont. the first is known as supervised machine learning and the last is knows as unsupervised machine learning. 

for this series we are going to focus on supervised machine learning. 

Ok and so supervised machine learning algorithms can be broken down even further. In to regression algorithms and classification algorithms. 

Regression algorithms: are used to predict a continuous output - outputs that can span an interval. weights, stock prices and fantasy football are all continuous outputs and thus can be predicted using regression algorithms

Simple Linear Regression or (SLR) the way slr works is it takes in one explanatory variable, also known as an independent variable and outputs continous value. In this case that will be used to predict fantasy football performance

The first thing we need to do before using the slr to predeict fantasy points is to train the algorithm. so we split our data in to training data and testing data. we will use a function to do this in the code, and we take a algorithm and input our training data into it. it takes this data to predict fantasy points in the future.from there we the evaluate how the predicted values faired with the actual values and calculate the ROOT MEAN SQUARED ERROR- this tells us how our model did. 

First thing we need to do is seperate the training and testing data. its not a good idea to test the model on the same data you used to train it. 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df['Usage'].values
Y = df['FantasyPoints'].values
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

So what happened here?

First we imported sklearn library called train_test_split tp split our data in to training and testing sets. sklearn is provided by scikit-learn the defacto machine learning library for python 

we set our X and Y variables, 'Usage' is our X as its our independent variable- its being used to test our theory for fantasy performance. Y is set to our fantasy points column as its being explained and predicted by our model. 

by default, train_test)split needs a numpy array as arguments to its function. the way we do this in pandas is by tacking on the  .values at the end there. Pandas then takes our dataframe column or series and converts it to a numpy array 

Soooo.... what does test_size=0.2 mean? its telling skilearn here that we'd like to save 20% of our data for testing & the remaining 80% will be used in our algorithm

train_test_split gives us back 4 values by default. In Python, when a function gives us back multiple values we can "unpack" the return values of the function using this notation or syntax.
